## Sample Data

This sample data is from ISLR. It displays sales (in thousands of units) for a particular product as a function of advertising budgets (in thousands of dollars) for TV, radio, and newspaper media.

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

Examining relationship between each advertising source

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

relationships between features -  plot multiple pairwise bivariate distributions in a dataset

```python
sns.pairplot(df,diag_kind='kde')
```

# Intro to Scikit Learning
Scikit-learn (Sklearn) is the most useful and robust library for machine learning in Python. It provides a selection of efficient tools for machine learning and statistical modeling including classification, regression, clustering and dimensionality reduction via a consistence interface in Python.

```python
X = df.drop('sales',axis=1) # everything except sales
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

Random State:
train_test_split splits arrays or matrices into random train and test subsets.

```python
# random_state:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```
